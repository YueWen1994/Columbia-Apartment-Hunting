```python
# Please find new york in order to see the map cuz some lat and long for apt are in other states
# I did not continue on Jeff 's work cuz Jeff drops some data from excel which affect the index for the loop of my method
# not sure if you guys like the color, you can change it if you want
# There are some problems with the raw data.I cleaned my data in the excel before running the code. 1. price is the string 2. quite a lot
#N.A for latitude and longitude. Also, longitude and latitude are incorrect. Some are in the other states. Not sure why.
#Here I use the price/bedroom to get the price color, you guys can inculde bathroom if interested.

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mp
import datetime as dt
import pandas as pd
import re
import gmplot
import numpy as np



df = pd.read_excel("all_data_final.xlsx")
price = df['price/$']
bedroom = df['bdrms']
latitudes = df["latitude"]
longitudes = df["longitude"]

lat1,lat2,lat3,lat4,lat5 = [],[],[],[],[]
lon1,lon2,lon3,lon4,lon5 = [],[],[],[],[]
ppd=[]
for i in range(0,len(latitudes)):
    if bedroom[i] == 1:
        p = int(price[i])
    elif int(bedroom[i])==0:
        p = int(price[i]) 
    else:
        p = int(price[i])/int(bedroom[i])
    ppd.append(p)
    if p < 1200:
        lat1.append(latitudes[i])
        lon1.append(longitudes[i])
    elif p < 1500:
        lat2.append(latitudes[i])
        lon2.append(longitudes[i])
    elif p < 1900:
        lat3.append(latitudes[i])
        lon3.append(longitudes[i])
    else:
        lat4.append(latitudes[i])
        lon4.append(longitudes[i])
        
gmap = gmplot.GoogleMapPlotter(latitudes.mean(),longitudes.mean(),15)
gmap.scatter(lat1, lon1,size=30,color='#00ff00',marker=False)
gmap.scatter(lat2, lon2,size=30,color='#FFFF00',marker=False)
gmap.scatter(lat3, lon3,size=30,color='#FFA500',marker=False)
gmap.scatter(lat4, lon4,size=30,color='#ff0000',marker=False)
gmap.draw('test.html')
```
