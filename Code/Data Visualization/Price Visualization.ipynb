```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mp
import datetime as dt
import pandas as pd
import re
import gmplot
import seaborn as sns
```

```python
df = pd.read_excel("all_data_final.xlsx")
```

```python
price = df[['price/$','bdrms']]
```

```python
sns.jointplot(df['price/$'],df['bdrms'])
```

```python
price.plot.hist(y='price/$').set_xlabel('Prices Range in $')
```

```python
price.plot.scatter(x='price/$',y='bdrms')
```

```python
price.plot.hist(y='bdrms',bins=5,color='green').set_xlabel('Bedrooms')
```

```python
price2 = price
```

```python
for i in range(len(price2)):
    try:
        price2.ix[i]['price/$']=price2.ix[i]['price/$']/price2.loc[i]['bdrms']
    except:
        continue
```

```python
price2.plot.hist(y='price/$').set_xlabel('Prices Range in $ per bedroom')
```
